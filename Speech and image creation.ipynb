```python
import os
import wave
from google import genai
from google.genai import types

# Yahan apni uploaded JSON file ka exact naam daalein
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/content/the-reesh-digest-tts-d24497f2e022.json"

# Apna exact Project ID yahan daalein (Cloud Console ke Dashboard se dekh kar)
PROJECT_ID = "the-reesh-digest-tts" # Isko apne ID se replace karna mat bhoolna
LOCATION = "us-central1"

# Vertex AI Client Initialize (Ab yeh direct JSON key se chupchaap login kar lega)
client = genai.Client(vertexai=True, project=PROJECT_ID, location=LOCATION)

script_text = """
W H O ne admit kiya hai ki wo Iran mein nuclear catastrophe ke liye taiyari kar raha hai.
Bushehr Nuclear Plant ke kafi kareeb missile girne ke baad khatra kaafi badh gaya hai.
W H O ke mutabiq... agar nuclear weapons ka istemal hua, toh tabahi ko rokna namumkin hoga.
Kya hum ek badi parmanu jung ke kagar par hain?
Aapka kya sochna hai? Comment karke bataiye... aur SUBSCRIBE kariye!
"""

style_instruction = "Say the following like a serious, suspenseful geopolitical news anchor. Make it sound dramatic and urgent. Emphasize the words 'catastrophe' and 'namumkin'."

print("Generating Pro Audio with JSON Auth...")

full_prompt = f"{style_instruction}\n\n{script_text}"

# Ekdum EXACT screenshot wali settings
response = client.models.generate_content(
    model="gemini-2.5-pro-tts",
    contents=full_prompt,
    config=types.GenerateContentConfig(
        speech_config=types.SpeechConfig(
            language_code="hi-IN",
            voice_config=types.VoiceConfig(
                prebuilt_voice_config=types.PrebuiltVoiceConfig(
                    voice_name="Zubenelgenubi"
                )
            )
        )
    )
)

audio_data = response.candidates[0].content.parts[0].inline_data.data

output_filename = "The_Reesh_Digest_Final_Audio.wav"
with wave.open(output_filename, "wb") as f:
    f.setnchannels(1)
    f.setsampwidth(2)
    f.setframerate(22050)
    f.writeframes(audio_data)

print(f"✅ Success! Audio saved perfectly as {output_filename}")
```

```python
import os
import time  # Naya module time delay ke liye
import vertexai
from vertexai.preview.vision_models import ImageGenerationModel

os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/content/the-reesh-digest-tts-d24497f2e022.json"
PROJECT_ID = "the-reesh-digest-tts" # Apna actual Project ID yahan daalein
LOCATION = "us-central1"

print("Initializing Vertex AI...")
vertexai.init(project=PROJECT_ID, location=LOCATION)

model = ImageGenerationModel.from_pretrained("imagen-3.0-generate-001")

# Prompts thode "PG-13" kar diye hain taaki Google block na kare
prompts = {
    "01_WHO_Alert": 'A hyper-realistic news broadcast graphic. Background shows a map of Iran with glowing nuclear plant icons. In the foreground, a World Health Organization (WHO) official logo. A large, bold neon text overlay reads: "WHO ALERT: NUCLEAR PREP!". Cinematic lighting, dramatic atmosphere.',

    "02_Bushehr_Target": 'A dramatic, photorealistic view of the Bushehr Nuclear Power Plant at dusk. A glowing red digital alert overlay is blinking on the facility. Intense news style. A bold text overlay reads: "BUSHEHR PLANT CRISIS?". High resolution.',

    "03_Catastrophe": 'A photorealistic image of a glowing, hazardous radiation symbol over a dark landscape. Intense glowing radiation. In the foreground, a sleek news quote graphic box with the text: "IMPOSSIBLE TO STOP CATASTROPHE". 8k resolution, highly detailed, dramatic.',

    "04_Nuclear_War": 'A high-tech digital global map in a military war room. The countries Iran, USA, and Israel are highlighted in glowing red. Red conflict lines connect them. A large glowing 3D text overlay asks: "GLOBAL CONFLICT?". Futuristic, strategic aesthetic.',

    "05_Subscribe_Outro": 'A stylish, high-tech background for a YouTube Shorts outro. A dark global map with neon accents. A glowing 3D animated "SUBSCRIBE" button and a notification bell icon floating in the center. Cinematic lighting.'
}

print("\n🚀 Starting Image Generation with Quota Management (thoda time lagega)...\n")

for scene_name, prompt in prompts.items():
    print(f"Generating scene: {scene_name}...")
    try:
        images = model.generate_images(
            prompt=prompt,
            number_of_images=1,
            language="en",
            aspect_ratio="9:16"
        )

        # Check karna ki Google ne safety ke wajah se khali result toh nahi bheja
        if images:
            file_name = f"{scene_name}.png"
            images[0].save(location=file_name)
            print(f"✅ Saved visually stunning image: {file_name}")
        else:
            print(f"⚠️ Safety filter blocked {scene_name}. Try modifying prompt.")

    except Exception as e:
        print(f"❌ Error generating {scene_name}: {e}")

    # Quota bachaane ke liye 30 seconds ka break (Aakhri image ke baad wait nahi karna)
    if scene_name != list(prompts.keys())[-1]:
        print("⏳ Waiting 30 seconds for quota reset...\n")
        time.sleep(30)

print("\n🎉 All done! Aapke saare premium visuals ab ready hone chahiye.")
```
