```python
import os
import wave
from google import genai
from google.genai import types

# Yahan apni uploaded JSON file ka exact naam daalein
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/content/the-reesh-digest-tts-d24497f2e022.json"

# Apna exact Project ID yahan daalein (Cloud Console ke Dashboard se dekh kar)
PROJECT_ID = "the-reesh-digest-tts" # Isko apne ID se replace karna mat bhoolna
LOCATION = "us-central1"

# Vertex AI Client Initialize (Ab yeh direct JSON key se chupchaap login kar lega)
client = genai.Client(vertexai=True, project=PROJECT_ID, location=LOCATION)

script_text = """
W H O ne admit kiya hai ki wo Iran mein nuclear catastrophe ke liye taiyari kar raha hai.
Bushehr Nuclear Plant ke kafi kareeb missile girne ke baad khatra kaafi badh gaya hai.
W H O ke mutabiq... agar nuclear weapons ka istemal hua, toh tabahi ko rokna namumkin hoga.
Kya hum ek badi parmanu jung ke kagar par hain?
Aapka kya sochna hai? Comment karke bataiye... aur SUBSCRIBE kariye!
"""

style_instruction = "Say the following like a serious, suspenseful geopolitical news anchor. Make it sound dramatic and urgent. Emphasize the words 'catastrophe' and 'namumkin'."

print("Generating Pro Audio with JSON Auth...")

full_prompt = f"{style_instruction}\n\n{script_text}"

# Ekdum EXACT screenshot wali settings
response = client.models.generate_content(
    model="gemini-2.5-pro-tts",
    contents=full_prompt,
    config=types.GenerateContentConfig(
        speech_config=types.SpeechConfig(
            language_code="hi-IN",
            voice_config=types.VoiceConfig(
                prebuilt_voice_config=types.PrebuiltVoiceConfig(
                    voice_name="Zubenelgenubi"
                )
            )
        )
    )
)

audio_data = response.candidates[0].content.parts[0].inline_data.data

output_filename = "The_Reesh_Digest_Final_Audio.wav"
with wave.open(output_filename, "wb") as f:
    f.setnchannels(1)
    f.setsampwidth(2)
    f.setframerate(22050)
    f.writeframes(audio_data)

print(f"✅ Success! Audio saved perfectly as {output_filename}")
```
